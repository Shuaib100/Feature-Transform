# Normalization/MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd 
df = pd.read_csv('agora.csv')
df.head()
```

```python
n = MinMaxScaler()
```

```python
n_fit= n.fit(df[['Marketing Spend']])
n_fit
```

```python
newn = df.drop(['Area'],axis = 1) # row = 0 , column = 1
newn.head()

```

```python
newn['Marketing Spend'] = n.transform(newn[['Marketing Spend']]) # if more then one word
newn.head()
```

```python
newn.Administration= n.fit_transform(newn[['Administration']]) # only for one word
newn.head()
```

```python
newn['Transport']= n.fit_transform(newn[['Transport']])
newn.head()
```

```python
newn['Profit']= n.fit_transform(newn[['Profit']])
newn.head()
```

# Normalization in one line

```python
newn[['Marketing Spend','Administration','Transport','Profit']] = n.fit_transform(newn[['Marketing Spend','Administration','Transport','Profit']])
newn.head()
```

# Array from normalized dataset 

```python
new_n = n.fit_transform(newn)
new_n
```

# Normalization from array 

```python
Data_n = pd.DataFrame(new_n)
Data_n.head()
```

```python
Data_n = pd.DataFrame(new_n, columns= ['Marketing Spend','Administration','Transport','Profit'])
Data_n.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
s = StandardScaler()
s
```
